```python
import kagglehub
import os
from PIL import Image

try:
    path = kagglehub.dataset_download("moltean/fruits")
except Exception:
    path = "/Users/stevensung/.cache/kagglehub/datasets/moltean/fruits/versions/11"

print("Path to dataset files:", path)
```

```python
from tqdm import tqdm

train_image_folder = os.path.join(path, 'fruits-360_dataset_100x100', 'fruits-360', 'Training')
train_images = []

for root, dirs, files in tqdm(os.walk(train_image_folder), desc="Loading images"):
    for file in files:
        if file.endswith('.jpg') or file.endswith('.png'):
            img_path = os.path.join(root, file)
            img = Image.open(img_path)
            label = root.split('/')[-1][:-2]
            train_images.append((img, label))

print(f"Total images loaded: {len(train_images)}")
```

```python
from matplotlib import pyplot as plt
import random

# Select 9 random images
random_images = random.sample(train_images, 8)

# Create a figure to display the images
fig, axes = plt.subplots(2, 4, figsize=(7.5, 7.5))

# Display each image with its label
for ax, (img, label) in zip(axes.flatten(), random_images):
    ax.imshow(img)
    ax.set_title(label)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
